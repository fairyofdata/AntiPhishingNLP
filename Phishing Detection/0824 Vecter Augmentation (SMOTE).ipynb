```python
import nltk
from nltk.corpus import wordnet
from random import randint
from tqdm.notebook import tqdm

# 필요한 nltk 패키지를 다운로드합니다. (로컬에서 실행시 필요)
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

def synonym_replacement(words, n=5):
    """
    단어 리스트를 받아서 n개의 단어를 동의어로 교체하여 새로운 단어 리스트를 반환
    """
    new_words = words.copy()
    random_word_list = list(set([word for word in words if word]))
    random_synonym_list = []

    # 동의어 찾기
    for random_word in random_word_list:
        synonyms = get_synonyms(random_word)
        if synonyms:
            random_synonym_list.append(synonyms)

    # n개의 단어를 동의어로 교체
    for _ in range(n):
        if random_synonym_list:
            random_synonym = random_synonym_list[randint(0, len(random_synonym_list)-1)]
            synonym = random_synonym[0]
            new_words = [synonym if word == random_synonym[1] else word for word in new_words]

    return new_words

def get_synonyms(word):
    """
    주어진 단어의 동의어를 반환
    """
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonym = lemma.name().replace("_", " ").replace("-", " ").lower()
            synonyms.add(synonym)
    if word in synonyms:
        synonyms.remove(word)
    return list(synonyms)

def augment_sentence(sentence, num_new_sentences=5):
    """
    주어진 문장을 바탕으로 num_new_sentences 개의 새로운 문장을 생성
    """
    new_sentences = []
    words = sentence.split(' ')
    for _ in range(num_new_sentences):
        new_sentence = synonym_replacement(words)
        new_sentences.append(' '.join(new_sentence))
    return new_sentences

# 데이터 확장
augmented_texts = []
for sentence in tqdm(voise_df['Sentence'], desc="Augmenting Data"):
    augmented_texts.extend(augment_sentence(sentence))

# 결과 확인
augmented_texts[:10]
```

### text-davinci-002 모델을 이용해 동의어 치환 수행 (거의 4천원이나 썼음) ㅠㅠ

```python
import random
import re

def find_synonyms(word):
    response = openai.Completion.create(
      engine="text-davinci-002",
      prompt=f"Give me synonyms for the word '{word}'",
      max_tokens=50
    )
    synonyms = response.choices[0].text.strip().split(', ')
    return synonyms if synonyms else [word]

def generate_alternative_sentences_refined_v2(sentence, morphemes):
    target_pos = ['NNG', 'VV', 'VA', 'NNP']
    target_words = [morph[0] for morph in morphemes if morph[1] in target_pos]
    new_sentences = set([sentence])  # 중복을 피하기 위해 집합 사용
    max_attempts = 20
    attempts = 0

    while len(new_sentences) < 5 and attempts < max_attempts:
        new_sentence = sentence
        words_to_replace = random.sample(target_words, min(2, len(target_words)))

        for word in words_to_replace:
            synonyms = find_synonyms(word)
            valid_synonyms = [syn for syn in synonyms if syn != word]

            if valid_synonyms:  # 유효한 동의어가 있을 경우
                synonym = random.choice(valid_synonyms)
                new_sentence = new_sentence.replace(word, synonym, 1)  # 처음 나오는 단어만 교체
            else:
                break

        if new_sentence not in new_sentences:
            new_sentences.add(new_sentence)

        attempts += 1

    return list(new_sentences)


# 문자열 형태의 형태소 데이터를 파싱하는 함수를 다시 정의합니다.
def parse_morphemes(morpheme_str):
    tokens = re.findall(r"Token\(form='(.*?)', tag='(.*?)', start=(\d+), len=(\d+)\)", morpheme_str)
    return [(form, tag, int(start), int(length)) for form, tag, start, length in tokens]


# 동의어 치환 함수를 형태소 데이터에 적용
all_sentences_morphemes_refined_v2 = []
total_rows = len(voise_antinoise_morphemes)
for idx, row in voise_antinoise_morphemes.iterrows():
    sentence = row['Sentence']
    morphemes = parse_morphemes(row['Morphemes'])
    all_sentences_morphemes_refined_v2.extend(generate_alternative_sentences_refined_v2(sentence, morphemes))
    print(f"진행 상황: {idx+1}/{total_rows} ({(idx+1)/total_rows*100:.2f}%)")

# 결과 출력
len(all_sentences_morphemes_refined_v2), all_sentences_morphemes_refined_v2[:10]  # 처음 10개 문장만 출력
```

```python
from transformers import BertTokenizer, BertModel
from tqdm import tqdm

# KoBERT 모델과 토크나이저를 로드합니다.
tokenizer = BertTokenizer.from_pretrained('monologg/kobert')
model = BertModel.from_pretrained('monologg/kobert')

def get_kobert_embedding(sentence):
    # 문장을 KoBERT의 입력 형식에 맞게 변환합니다.
    inputs = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True, max_length=128)

    # 임베딩을 추출합니다.
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs['last_hidden_state'][:,0,:].numpy()  # [CLS] 토큰의 임베딩만 반환

# normalcall의 모든 문장에 대해 임베딩을 수행합니다.
embedded_sentences_normal = [get_kobert_embedding(sentence) for sentence in tqdm(normalcall['text'], desc="임베딩 진행중", ncols=100)]

import numpy as np

# 임베딩 결과를 numpy 배열로 변환합니다.
embedded_array = np.array(embedded_sentences_normal)

# 임베딩 결과를 파일로 저장합니다.
np.save('/content/drive/MyDrive/0824_embedded_sentences_normal.npy', embedded_array)
```

```python
import numpy as np
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE

# 데이터 로드
embedded_fish = np.load('/content/drive/MyDrive/0824_embedded_sentences.npy')
embedded_norm = np.load('/content/drive/MyDrive/0824_embedded_sentences_normal.npy')

# 라벨 생성
labels_fish = np.ones((embedded_fish.shape[0],))
labels_norm = np.zeros((embedded_norm.shape[0],))

# 데이터의 형태를 2차원으로 변환
embedded_fish_2d = embedded_fish.reshape(embedded_fish.shape[0], -1)
embedded_norm_2d = embedded_norm.reshape(embedded_norm.shape[0], -1)

# 데이터와 라벨 합치기
X_combined = np.vstack((embedded_fish_2d, embedded_norm_2d))
y_combined = np.hstack((labels_fish, labels_norm))

# SMOTE 적용
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_combined, y_combined)
```
