```python
import os
import json
import pandas as pd

# 디렉토리 내의 모든 파일 목록 가져오기
directory_path = "/content/drive/MyDrive/normalcall/04.배송"
all_files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]

# 파일 목록을 200개 단위로 분할
file_batches = [all_files[i:i+200] for i in range(0, len(all_files), 200)]

# 결과를 저장할 데이터프레임 초기화
df_texts = pd.DataFrame(columns=["text"])

# 오류가 발생한 파일을 저장할 리스트 초기화
error_files = []

# 각 파일 배치에 대해 처리
for idx, batch in enumerate(file_batches):
    for file in batch:
        try:
            # JSON 파일 불러오기
            with open(os.path.join(directory_path, file), "r", encoding="utf-8") as json_file:
                data = json.load(json_file)
                # 'info' 안의 'text' 필드 추출
                batch_texts = [item['annotations']['text'] for item in data['info'] if 'annotations' in item and 'text' in item['annotations']]
                # 데이터프레임에 추가
                df_texts = df_texts.append(pd.DataFrame(batch_texts, columns=["text"]), ignore_index=True)
        except Exception as e:
            # 오류가 발생한 파일명 저장
            error_files.append(file)

    # 200개 작업이 끝난 후, 다음 100개 파일을 처리하기 전에 현재 상태 출력
    if idx * 200 < len(all_files):
        print(f"{idx * 200}개의 파일 처리 완료")

# 오류가 발생한 파일 목록 출력
error_files

# 처음 5개의 데이터를 출력하여 확인
df_texts.head()
```

```python
import os
import json
import pandas as pd
from tqdm import tqdm

# 결과를 저장할 데이터프레임 리스트 초기화
df_list = []

# /content/drive/MyDrive/normalcall 디렉토리의 하위 디렉토리 탐색
base_directory = "/content/drive/MyDrive/normalcall"
subdirectories = [d for d in os.listdir(base_directory) if os.path.isdir(os.path.join(base_directory, d))]

# 각 하위 디렉토리에 대해 처리
for subdirectory in tqdm(subdirectories, desc="Processing subdirectories"):
    current_directory = os.path.join(base_directory, subdirectory)
    all_files = [f for f in os.listdir(current_directory) if os.path.isfile(os.path.join(current_directory, f))]

    # 파일 목록을 200개 단위로 분할
    file_batches = [all_files[i:i+200] for i in range(0, len(all_files), 200)]

    # 결과를 저장할 데이터프레임 초기화
    df_sub_texts = []

    # 각 파일 배치에 대해 처리
    for batch in tqdm(file_batches, desc=f"Processing files in {subdirectory}"):
        batch_data = []
        for file in batch:
            try:
                # JSON 파일 불러오기
                with open(os.path.join(current_directory, file), "r", encoding="utf-8") as json_file:
                    data = json.load(json_file)
                    # 'info' 안의 'text' 필드 추출
                    batch_texts = [item['annotations']['text'] for item in data['info'] if 'annotations' in item and 'text' in item['annotations']]
                    batch_data.extend(batch_texts)
            except Exception as e:
                # 오류가 발생하면 해당 파일명과 오류 메시지를 출력
                print(f"Error in file {file}: {str(e)}")

        # 데이터프레임에 추가
        df_sub_texts.append(pd.DataFrame(batch_data, columns=["text"]))

    # 해당 하위 디렉토리의 데이터프레임을 리스트에 추가
    df_list.append(pd.concat(df_sub_texts, ignore_index=True))

# 모든 데이터프레임을 하나로 합치기
final_df = pd.concat(df_list, ignore_index=True)

# 결과 데이터프레임의 처음 5개 데이터 확인
print(final_df.head())
```
