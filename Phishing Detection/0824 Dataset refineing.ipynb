```python
import pandas as pd

# JSON 파일 읽기
file_path = '/content/drive/MyDrive/주거와생활.json'
dfa = pd.read_json(file_path)

# DataFrame 출력
print(dfa)
```

```python
print(dfa['data'].iloc[5])
'utterance'
```

```python
# 각 대화 데이터에 대해 작업을 수행하는 함수 정의 (예외 처리 추가)
def process_dialogue_without_speaker_modified(dialogue_data):
    dfa = pd.DataFrame(dialogue_data['body'])
    if 'utterance' not in dfa.columns:
        return ''  # 'utterance' 항목이 없으면 빈 문자열 반환
    return ' '.join(dfa['utterance'])

# 결과를 저장할 데이터프레임 리스트
result_texts_with_exception_handling = []

# 각 행의 'data' 컬럼에 함수 적용
for index, row in dfa.iterrows():
    processed_text = process_dialogue_without_speaker_modified(row['data'])
    result_texts_with_exception_handling.append(processed_text)
    print(f"{index + 1}번째 열이 처리되었습니다. 처리된 텍스트: '{processed_text}'\n")

# 결과 확인 (첫 번째 대화 데이터만 출력)
result_texts_with_exception_handling[0]
```

```python
import re

def refine_text_corrected(text):
    # #으로 시작해서 #으로 끝나는 모든 단어 제거
    text = re.sub(r'#.*?#', '', text)

    # 마침표, 쉼표, 느낌표, 물음표, 큰따옴표, 작은따옴표를 제외한 모든 특수문자 제거
    text = re.sub(r'[^가-힣0-9a-zA-Z.,!?“”‘’\s]', '', text)

    # 4회 이상 똑같은 글자가 연속으로 나오는 경우 3개만 남기고 제거
    text = re.sub(r'(.)\1{3,}', r'\1\1\1', text)

    return text

# 위의 함수를 result_texts_with_exception_handling 리스트의 각 항목에 적용
refined_texts_corrected = [refine_text_corrected(text) for text in result_texts_with_exception_handling]

# 결과 확인 (첫 번째 데이터만 출력)
refined_texts_corrected[197387:197399]
```

```python
# refined_texts_corrected 리스트를 DataFrame으로 변환
refined_df = pd.DataFrame(refined_texts_corrected, columns=['refined_text'])

refined_df.head()  # 처음 5개 데이터를 출력하여 확인
refined_df.to_csv()
```
